Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location.
You have to scrape the job-title, job-location, company_name, experience_required. You have to
scrape first 10 jobs data.
This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.
3. Then click the searchbutton.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.shine.com/")
```

```python
designation = driver.find_element(By.CLASS_NAME,"form-control  ")
designation.send_keys("Data Analyst")
```

```python
location = driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div[1]/form/div/div[1]/ul/li[2]/div/input")
location.send_keys("Banglore")
```

```python
search = driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div[1]/form/div/div[2]/div/button")
search.click()
```

```python
job_titles = []
job_locations = []
company_names = []
experience_required = []
```

```python
title = driver.find_elements(By.XPATH,'//strong[@class="jobCard_pReplaceH2__xWmHg"]')[0:20]
for i in title:
    title_tag=i.text
    job_titles.append(title_tag)

location = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')[0:20]
for i in location:
    location_tag=i.text
    job_locations.append(location_tag)

names = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')[0:20]
for i in names:
    name_tag=i.text
    company_names.append(name_tag)

experience = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')[0:20]
for i in experience:
    experience_tag=i.text
    experience_required.append(experience_tag)
```

```python
print(len(job_locations),len(job_titles),len(company_names),len(experience_required))
```

```python
import pandas as pd
df=pd.DataFrame({'title':job_titles,'location':job_locations, 'company_name':company_names,'experience_required':experience_required})
df
```

To scrape the data you have to go through following steps:
1. Go to Flipkart webpage by url : https://www.flipkart.com/
2. Enter “sunglasses” in the search fieldwhere “search for products, brands and more” is written and
click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the
required data as usual.
4. After scraping data from the first page, go to the “Next” Button at the bottom other page , then
click on it.
5. Now scrape data from this page as usual
6. Repeat this until you get data for 100sunglasses. Note: That all of the above steps have to be done
by coding only and not manually.

```python
import selenium
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.flipkart.com/")
```

```python
sunglasses=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input")
sunglasses.send_keys("sunglasses")
```

```python
search=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button")
search.click()
```

```python
Brand = []
description = []
price = []
```

```python
start=0
end =5
for page in range(start,end):
    
    Brands=driver.find_elements(By.XPATH,'//div[@class="syl9yP"]')
    for i in Brands:
        Brand_tag=i.text
        Brand.append(Brand_tag)
    
    prices=driver.find_elements(By.XPATH,'//div[@class="Nx9bqj"]')
    for i in prices:
        price_tag=i.text
        price.append(price_tag)
    
    descriptions=driver.find_elements(By.XPATH,'//a[@class="WKTcLC"]')
    for i in descriptions:
        description_tag=i.text
        description.append(description_tag)
        
    next_button= driver.find_element(By.XPATH,'//a[@class="_9QVEpD"]')
    next_button.click()
    time.sleep(3)



```

```python
print(len(Brand),len(price),len(description))
```

```python
import pandas as pd
df=pd.DataFrame({'title':Brand,'Price':price})
df
```
