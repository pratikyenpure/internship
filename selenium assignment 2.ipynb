Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped:
“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”. 

```python
def scrape_billionaires():
    driver = webdriver.Chrome()
    driver.get("https://www.forbes.com/billionaires/")
    
    

    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    

    billionaires = []
    rows = driver.find_elements(By.XPATH, "//div[@class='table-row']")
    for row in rows:
        details = {}
        details["Rank"] = row.find_element(By.CLASS_NAME, "rank").text
        details["Name"] = row.find_element(By.CLASS_NAME, "personName").text
        details["Net worth"] = row.find_element(By.CLASS_NAME, "netWorth").text
        details["Age"] = row.find_element(By.CLASS_NAME, "age").text
        details["Citizenship"] = row.find_element(By.CLASS_NAME, "countryOfCitizenship").text
        details["Source"] = row.find_element(By.CLASS_NAME, "source-column").text
        details["Industry"] = row.find_element(By.CLASS_NAME, "category").text
        billionaires.append(details)
    
   

billionaires_list = scrape_billionaires()
for billionaire in billionaires_list:
    print(billionaire)
```
